```python
import pandas as pd
```

```python
csv1 = pd.read_csv('michelin_hotels.csv',index_col=0)
df1 = pd.DataFrame(csv1)
df_left = df1[df1["Hotel_Name"].str.contains("Atlanta")==False]
len(df_left)
```

```python
df_left.head(10)
```

```python
csv2 = pd.read_csv('not_atlantean.csv',index_col=0)
df2 = pd.DataFrame(csv2)
df2.head()
```

```python
df_with_null = pd.concat([df_left,df2])
```

```python
len(df_with_null)
```

```python
df_final = df_with_null[df_with_null["Hotel_Name"].str.contains("NULL")==False]
```

```python
len(df_final)
```

```python
df_final
```

```python
df_final = df_final.loc[:, ~df_final.columns.str.contains('^Unnamed')]
```

```python
df_final.head()
```

```python
df_final.to_csv('Michelin_Data_Final.csv')
```
