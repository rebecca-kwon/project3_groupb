```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load in csv download from original source
michelin_star = pd.read_csv('./Michelin_Source/michelin_star.csv')
michelin_star.head()
```

```python
# check number of unique cities
michelin_cities = michelin_star.groupby('city')['name'].nunique()
print(michelin_cities)
```

```python
# check number of unique cuisine types
michelin_cuisines = michelin_star.groupby('cuisine1')['name'].nunique()
print(michelin_cuisines)
```

```python
# check the columns
michelin_star.columns
```

```python
# drop blurb colum
michelin_star = michelin_star.drop('blurb', axis=1)
michelin_star.head()
```

```python
# export csv
michelin_star.to_csv('./Michelin_Source/michelin_clean.csv', index = False)
```
